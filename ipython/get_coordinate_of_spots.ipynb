```python
from math import sin, asin, cos, radians, fabs, sqrt

def haversine(lon1, lat1, lon2, lat2): # 经度1，纬度1，经度2，纬度2 （十进制度数）
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # 将十进制度数转化为弧度
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    # haversine公式
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371 # 地球平均半径，单位为公里
    return c * r * 1000


# 38352.523132 (m为单位)
print(haversine(113.973129, 22.599578, 114.3311032, 22.6986848))
```

```python
#calcuate each station, near r=500m college/hospital/train station

distance_list = []
#calcuate the distance between each spot and station
for spot in df.iterrows():
    distance_dict = {}
    spot_lat = spot[1]['latitude']
    spot_lng = spot[1]['longtitude']
    print('spot:',spot[0],', lat:',spot_lat,' lng:',spot_lng)
    distance_dict['spot_coor'] = [spot_lat,spot_lng]
    distance_dict['spot'] = spot[0]
    station_list = []
    for station in sta_info_df.iterrows():
        sdist_dict = {}
        sta_lat = station[1]['lat']
        sta_lng = station[1]['lon']
        sdist_dict['station_coor'] = [sta_lat,sta_lng]
        sdist_dict['station'] = station[1]['sno']
        sdist_dict['distance'] = haversine(sta_lng,sta_lat,spot_lng,spot_lat)
        print('station:',station[0],', lat:',sta_lat,' lng:',sta_lng)
        print('>>> distance:',sdist_dict['distance'])
        station_list = station_list + [sdist_dict]
    distance_dict['station_list'] = station_list
    distance_list = distance_list + [distance_dict]
```

```python
total_df = pd.DataFrame()
for spot in distance_list:
    #print(spot)
    new_spot_df = pd.DataFrame(spot['station_list'])
    new_spot_df['spot_id'] = spot['spot']
    new_spot_df['near_1km'] = False
    new_spot_df['near_1km'][new_spot_df.distance<1000] = True
    new_spot_df['near_500m'] = False
    new_spot_df['near_500m'][new_spot_df.distance<500] = True
    
    total_df = total_df.append(new_spot_df)
    #name = "csv/spot/spot_1km_"+str(spot['spot']).zfill(2)+".csv"
    #new_spot_df.to_csv(name)
```
